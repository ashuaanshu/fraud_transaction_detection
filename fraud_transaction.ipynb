```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
a=pd.read_csv("creditcard.csv")
a.head(10)
```

```python
a.info()
```

```python
a.isnull().sum()
```

```python
a['Class'].value_counts()
```

```python
# this data is very unblanced
# 0 this is normal transaction
# 1 this is fraudulent transaction
```

```python
#seprating data for analysis
legit=a[a.Class==0]
fraud=a[a.Class==1]
```

```python
legit.shape
```

```python
#statistical measures of data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#campare the values for both transaction
a.groupby('Class').mean()
```

```python
# under-sampling
# Build a sample dataset containing similar distribution of normal transaction and fraudulent transcation
# number of fraudlent transaction -->492
```

```python
legit_Sample=legit.sample(n=492) #random data select
```

```python
#containing two datafram
```

```python
new_data=pd.concat([legit_Sample, fraud], axis=0)
```

```python
new_data.head(10)
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

```python
x=new_data.drop(columns='Class', axis=1)
y=new_data['Class']
```

```python
x
```

```python
y
```

```python
X_test, X_train, Y_test, Y_train= train_test_split(x,y,test_size=0.2, stratify=y, random_state=2)
```

```python
print(x.shape, X_train.shape, X_test.shape)
```

```python
#Model Traning
#logistic Regression
model=LogisticRegression()
model.fit(X_train, Y_train)
```

```python
X_train_predict=model.predict(X_train)
train_data_acc=accuracy_score(X_train_predict, Y_train)
```

```python
train_data_acc
```

```python
X_test_predict=model.predict(X_test)
test_data_acc=accuracy_score(X_test_predict, Y_test)
```
